II - Request to find networks in Montreal

```python
#Request for all stations in Paris
url = "http://api.citybik.es/v2/networks/bixi-montreal"
message = requests.get(url)
message = message.text
message = json.loads(message)
save = open("stations_in_Montreal.json", "w")  
json.dump(message, save, indent = 6) 
save.close()
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
#selecting data
message = pd.read_json('stations_in_Montreal.json')
try:
    if (type(message['network']['company']) == str):
        company_name = message['network']['company']
    else:
        company_name = message['network']['company'][0]
except:
    company_name = 'not available'

try:
    city_location = message['network']['location']['city']
except:
    city_location = 'not available'

latitude = []
for i in message['network']['stations']:
    try:
      latitude.insert(len(latitude), i['latitude'])
    except:
      latitude.insert(len(latitude), 0.0)

longitude = []
for i in message['network']['stations']:
    try:
      longitude.insert(len(longitude), i['longitude'])
    except:
      longitude.insert(len(longitude), 0.0)

nbr_of_bikes = []
for i in message['network']['stations']:
    try:
      nbr_of_bikes.insert(len(nbr_of_bikes), i['free_bikes'])
    except:
      nbr_of_bikes.insert(len(nbr_of_bikes), 0.0)

station_name = []
for i in message['network']['stations']:
    try:
      station_name.insert(len(station_name), i['name'])
    except:
      station_name.insert(len(station_name), 0.0)
    
#gathering the collected data
table = pd.DataFrame()
table['latitude'] = latitude
table['longitude'] = longitude
table['nbr of bikes'] = nbr_of_bikes
table['station name'] = station_name
table['city'] = city_location
table['company name'] = company_name

#check for duplicate
table = table.drop_duplicates(subset=['latitude', 'longitude'], keep='last')
table.head(5)
```

Put your parsed results into a DataFrame.

```python
#Saving result
table.to_csv('info_on_stations.csv', index=False)
table.head(5)
```
